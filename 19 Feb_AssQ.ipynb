**Q1.** What is Flask Framework? What are the advantages of Flask Framework?

**Ans:**

Flask is a popular micro web framework for building web applications using the Python programming language. Flask is classified as a "micro" framework because it doesn't require particular tools or libraries to be installed and can be easily extended to support more complex applications.

<b><u>Advantages</u></b>

1. Lightweight: Flask is a lightweight framework, making it easy to set up and get started with quickly.

2. Flexible:  Flask is flexible enough to accommodate a wide range of web application requirements. It's designed to be modular, so you can choose only the features you need.

3. Easy to learn: Flask is a simple and easy-to-learn framework, especially if you're already familiar with Python.

4. Large Community: Flask has a large and active community of developers, so it's easy to find help and support when you need it.

5. Extensible: Flask is extensible, meaning that you can add extensions or plugins to enhance its functionality.

6. Powerful: Flask provides a lot of functionality out of the box, such as support for web sockets, RESTful APIs, and more.

7. Scalable: Flask can be used to build small, simple web applications as well as large, complex web applications, so it's a scalable option for many projects.

**Q2.** Create a simple Flask application to display 'Hello World!!'. 

Attach the screenshot of the output in Jupyter Notebook.

**Ans:**

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "Hello, World!!"


if __name__=="__main__":
    app.run(host="0.0.0.0")
```

**Q3.** What is App routing in Flask? Why do we use app routes?

**Ans:**

App routing in Flask refers to the process of mapping a URL endpoint to a specific function in your Flask application. App routes are defined using the @app.route() decorator in Flask, which allows you to specify the URL path that the route should respond to and the function that should be called when that path is requested.

App routes are useful in Flask for several reasons:

1. Routing allows you to define how your application will respond to different requests, based on the URL path.

2. Routing helps you to organize your code and keep it modular, by separating different parts of your application into separate functions.

3. Routing makes it easy to add new features to your application by defining new routes and functions.

4. Routing allows you to create RESTful APIs, which can be used to build web services that other applications can consume.

**Q5.** What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

**Ans:**

The url_for() function is used for URL building. The url_for() function generates a URL to a specific function or endpoint, based on its name and any arguments that need to be passed to it. The url_for() function is a useful feature of Flask that makes it easy to generate URLs to specific endpoints in your application. It's especially useful when you have routes that take dynamic arguments, as it automatically handles URL encoding and other complexities.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    url = url_for('hello', name='welcome')
    return f'Hello URL: {url}'


@app.route('/hello/<name>')
def hello(name):
    return f'Hello, {name}!'

if __name__ == '__main__':
    app.run(host="0.0.0.0")
```

The index() function, which generates a URL for the hello() function, and the hello() function, which takes a name argument and returns a personalized greeting.

The url_for() function is used in the index() function to generate a URL to the hello() function, passing the name "John" as an argument. The resulting URL is then displayed in the response.
