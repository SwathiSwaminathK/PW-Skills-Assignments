**Q1.** Explain GET and POST methods.

**Ans:**

GET method: The GET method is used to retrieve data from a web server. When a GET request is made, the client requests data from the server and includes any necessary parameters in the URL. The server then returns the requested data as a response. GET requests are usually used to retrieve data, such as web pages, images, and other resources, from a server.

POST method: The POST method is used to submit data to a web server. When a POST request is made, the client sends data to the server as part of the request body, rather than as part of the URL. The server then processes the data and returns a response. POST requests are commonly used to submit form data, upload files, and perform other actions that involve changing or updating data on the server.

**Q2.** Why is request used in Flask?

**Ans:**

In Flask, the request object is used to represent an incoming HTTP request from a client to the Flask application. The request object contains all of the data and information associated with the request, such as the HTTP method (GET, POST, etc.), headers, cookies, and form data.

The request object is a fundamental component of web applications, as it allows you to retrieve and process data from incoming requests. 

You might use the request object to:

* Retrieve form data submitted by a user, such as their name, email address, and other information
* Access cookies stored on the client's machine, such as login credentials or session data
* Retrieve data from URL parameters or query strings
* Access information about the client, such as their IP address or user agent string
* Determine the HTTP method used to make the request, such as GET or POST

**Q3.** Why is redirect() used in Flask?

**Ans:**

In Flask, the redirect() function is used to redirect a user to a different URL. This is useful when you want to send a user to a different page after they've completed an action, such as submitting a form or logging in.

Here are some common scenarios where you might use the redirect() function:

* After a user submits a form, you might redirect them to a "thank you" page or a confirmation page.
* After a user logs in, you might redirect them to their account dashboard or the main page of your application.
* After a user completes a purchase, you might redirect them to a receipt page or a download page.

**Q4.** What are templates in Flask? Why is the render_template() function used?

**Ans:**

In Flask, templates are used to create HTML pages that can be dynamically generated and rendered using data from the application. Templates in Flask use a simple templating language called Jinja2, which allows you to insert variables, control structures (such as loops and conditionals), and other dynamic content into your HTML pages.

The render_template() function in Flask is used to render a template and return it as a response to an HTTP request. The render_template() function takes the name of the template as its first argument, and any variables that should be passed to the template as keyword arguments.

The render_template() function is a powerful tool in Flask, as it allows you to create dynamic, data-driven HTML pages without having to write the HTML by hand. By using templates and passing data to them as keyword arguments, you can create web applications that are flexible, maintainable, and easy to update.

**Q5.** Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.

**Ans:**

```python
from flask import Flask, request, jsonify

app = Flask(__name__)

@app.route('/greet', methods=['POST'])
def greet():
    name = request.json['name']
    message = 'Hello, {} !!!'.format(name)
    return jsonify({'message': message})

if __name__ == '__main__':
    app.run(host="0.0.0.0")
```
